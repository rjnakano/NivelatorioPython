```python
x = np.arange(0.01,1,0.01)
y1 = 1/x
y2 = 1/x**2
y3 = 1/x**3
y4 = 1/x**4

plt.plot(x,y1,'b.')
plt.plot(x,y2,'r--')
plt.plot(x,y3,'g-.')
plt.plot(x,y4,'g:')
plt.axis([0,1,0,100])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,'b.')
plt.subplot(2,2,2)
plt.plot(x,y2,'r--')
plt.subplot(2,2,3)
plt.plot(x,y3,'g-.')
plt.subplot(2,2,4)
plt.plot(x,y4,'g:')
plt.axis([0,1,0,100])
plt.show()
```

```python
import math

x = np.arange(-10,10,0.1)
ysin = np.sin(x)
ycos = np.cos(x)
ytan = np.tan(x)

yarcsin = np.arcsin(x)
yarccos = np.arccos(x)
yarctan = np.arctan(x)

plt.subplot(2,3,1)
plt.plot(x,ysin)

plt.subplot(2,3,2)
plt.plot(x,ycos)

plt.subplot(2,3,3)
plt.plot(x,ytan)

plt.subplot(2,3,4)
plt.plot(x,yarcsin)

plt.subplot(2,3,5)
plt.plot(x,yarccos)

plt.subplot(2,3,6)
plt.plot(x,yarctan)
```

```python
# Múltiples barras
N = 5
promedio_h = [20,35,30,35,27]
promedio_m = [25,32,34,20,25]

ind = np.arange(N)
ancho = 0.35
plt.bar(ind, promedio_h,ancho,label= 'Hombres')
plt.bar(ind+ancho, promedio_m,ancho,label= 'Mujeres')

plt.ylabel('Puntajes')
plt.title('Puntajes por Grupo y Género')

plt.xticks(ind + ancho/2,('G1','G2','G3','G4','G5'))
plt.legend()
plt.show()
```
